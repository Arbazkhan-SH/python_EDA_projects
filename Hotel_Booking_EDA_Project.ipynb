### Data description

**1. Hotel** : Resort hotel(H1) or city hotel(H2).

**2. is_cancelled** : Values indicating if the booking was cancelled (1) or not (0).

**3. lead_time** : Number of days that elapsed between the entering date of the booking into the PMS and the arrival date.

**4. arrival_date_year** : Year of arrival date.

**5. arrival_date_month** : Month of arrival date.

**6. arrival_date_week_number** : week number for arrival date.

**7. arrival_day_of_month** : Day of arrival date.

**8. stays_in_weekend_nights** : Number of weekend nights (Saturday or Sunday) the guest stayed or booked to stay at the hotel.

**9. stays_in_week_nights** : Number of week nights (Monday to Friday) the guest stayed or booked to stay at the hotel.

**10. adults** : Number of adults.

**11. children** : Number of children.

**12. babies** : Number of babies.

**13. meal** : Type of meal booked.(BB - Bed and Breakfast, FB - Full Board, HB - Half Board, SC - Self-Catering, undefined)


**14. country** : country of origin.

**15. market_segment** : Market segment designation. the term 'TA' means "Travel Agent" and 'TO' means "Tour Operators".

**16. distribution_Channel** : How the customer accessed to book stay corporate booking/Direcct/TA(Travel Agent),To(Tour Operators).

**17. is_repeated_guest** : values indicating if the booking name is repeated guest (1) or not (0).

**18. previous_cancellation** : Number of previous booking that were cancelled by the customer prior to the current booking.

**19. previous_booking_not_cancelled** : Number of previous bookings not cancelled by the customrer prior to the current booking.

**20. reserved_room_type** : Code of room type reserved.

**21. assigned_room_type** : Code for the room type assigned to the bookings.

**22. booking_changes** : Number of changes made to the booking from the moment thebooking was entered on the PMS until the moment of check-in or cancellation.

**23. deposit_type** : Indication on if the customer made a deposit to gruaratee the booking.

**24. agent** : ID of the travel agency that made the booking.

**25. company** : ID of the company that made the booking or responsible for paying the booking.

**26. days_in_waiting_list** : Number of days the booking was in waiting list before it was confirmed to the customer.

**27. customer_type** : Type of booking, assuming one of four categories.

**28. adr** : Average Daily Rate as defined by dividing the sum of all transaction by the total number of staying nights.

**29 required_car_parking_spaces** : Number of car parking space required by the customer.

**30 total_special_requests** Number of special request by the customer.

**31 reservation_status** : Reservation last status, assuming one of the three categories


*  Canceled - Booking was canceled by the customer
*  Check-out - Customer has checked in but already departed
*  No-show - Customer did not check-in and did inform the hotel of the reson why

**32 reservation_status_date** : Date at which the last status was set. The variable can be used in ci=onjunction withthe Reservation status to understand when eas the booking canceled or whendid the customer checked-out of the hotel.

##2. Which Agent made the more bookings ?

```python
# return the highest bookings by agent
highest_bookings = df1.groupby(['agent'])['agent'].agg({'count'}).reset_index().rename(columns={'count':'most_booking'}).sort_values(by='most_booking', ascending=False);

# removing the values which are equal to 0 in agent column
highest_bookings.drop(highest_bookings[highest_bookings['agent']==0].index,inplace=True);

#top 10 highest booking agents
top_10_highest_bookings = highest_bookings[:10]

print(top_10_highest_bookings)
```

```python
# plotting the bar cahrt for visualization
plt.figure(figsize=(15,10));
sns.barplot(x=top_10_highest_bookings['agent'],y=top_10_highest_bookings['most_booking'],order=top_10_highest_bookings['agent']);
plt.title('Number of Bookings by Agent');
plt.xlabel('Agent');
plt.ylabel('NUmber of bookings');


```

Agent ID no 9.0 has made highest number of bookings.

##10. Where the guests allocated with the same room type which has reserved by them ?

```python
#defining the function for the reserved room is same or not as assigned room type
def same_room_allotment(x):
  ''' function to check whether the reserved rooms are same or not as assigned room type '''

  if x['reserved_room_type'] != x['assigned_room_type']:
    return 1
  else:
    return 0;
```

```python
# crating a new column
df1['same_room_alloted_or_not'] = df1.apply(lambda x : same_room_allotment(x), axis=1)

#groupby is_canceled
groupby_canceled_df = df1.groupby('is_canceled')

#creating new variable of percentage of same room alloted or not
percentage_same_room_or_not_df = pd.DataFrame(groupby_canceled_df['same_room_alloted_or_not'].sum()*100/groupby_canceled_df.size()).rename(columns={0:'percentage'})

# printing percentage_same_room_or_not_df
print(percentage_same_room_or_not_df)

# Visualize with barplot

plt.figure(figsize=(10,8))
ax = sns.barplot(x=percentage_same_room_or_not_df.index, y=percentage_same_room_or_not_df['percentage'])
plt.title('percentage not getting the same room and booking cancelled')
plt.xlabel('is_canceled')
plt.ylabel('percentage of not getting same room as per reservation')
plt.xticks([0,1],['Not_cancelled','cancelled'])

for bars in ax.containers:
  ax.bar_label(bars)
```
