## Data Description

**Play Store Data description**



1.   Application Name : The name of the mobile application.
2.   Category : The primary category to which the app belongs (Games, Productivity, Education)
3.   Rating : The overall user rating of the app, represented as a numeric values.
4.   Reviews : The total number of users reviews for the app< as recorded when the data was scraped.
5.   Size : The size of the app file(e.g., 15MB, 50MB), as recorded when the data was scraped.
6.   Installs : The total number of times the app has been downloaded/installed by uses, as recordedwhen the data was scraped.
7.   Type : The cost of the app if it is paid, represented as a numeric values(e.g., $2.99). If the price is free, the price will be 0.

8.   Price : The cost of the app if it is paid, represent as a numeric value (e.g., $2.99). If the app is free, the price will be 0.
9.   Content Rating : The age group the app is targeted at (e.g., Everyone, Teen, Mature 21+, Adult).
10.   Genres : Additional categories or genres the app belongs to, apart from its main category(e.g., Musical, Family, Game).



**User Reviews Data Description**


1.   App : The name of the mobile application being reviwed.
2.   Translated_Review : The user review of the app, which has been preprocessed (e.g., cleaned, tokenized) and translated into english.
3.   Sentimental : The sentiment of the user review, Categorized as Positive, Negative, or Neutral, based on text analysis.
4.   Sentiment_Polarity : A numeric score representing the sentiment polarity of the review, ranging from -1 (most negative) to 1 (most positive).
5.   Sentiment_Subjectivity : A numeric score indicating the subjectivity of the review, ranging from 0 (objective) to (subjective), showing how subjective or opinion-based the review is.

### Defining the function to convert the KB to MB and then convert into foat datatype.

As we see the values of Size column are in different units M and K. K is stands fro KB and M is stands for MB. to analyse the data we need to covert differnt units into one unit, in this case we will convert all units into MB.



```python
# Function to convert the size string into fload (in MB)

def convert_size(size):
  if 'K' in size:
    return float(size.replace('K', '')) / 1024       # conver KB to MB
  elif 'M' in size:
    return float(size.replace('M', ''))      # Already in MB
  else:
    return size
```

```python
#n applying the function
ps_df['Size']= ps_df['Size'].apply(lambda x : convert_size(x))
```

```python
ps_df['Size'].isnull().sum()
```

### In installs column there are some unncessary characters like coma (,) and plus (+). which has to be remove to analyse the data.

```python
# Function to remove comma',' and plus'+' from the values of column Installs

def clean_installs(installs):
  return int(installs.replace(',','').replace('+',''))
```

```python
# Applying the function to column Installs
ps_df['Installs']=ps_df['Installs'].apply(clean_installs)
```

```python
# check unique values in the Price column
ps_df['Price'].unique()
```

As we can see that in Price column there is doller symbol. we need to remove that doller $ symbol to analyse.

```python
# Remove the $ symbol and convert the column to float
ps_df['Price'] = ps_df['Price'].str.replace('$','').astype(float)
```

```python
# converting last updated column into datetime
ps_df['Last Updated'] = ps_df['Last Updated'].apply(lambda x : datetime.strptime(x, '%B %d, %Y'))
```
