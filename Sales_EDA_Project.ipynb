### Exploratory data analysis

```python
# checking that which gender is purchased more
ax = sns.countplot(x='Gender', data=df)

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
# checking that who spent most amount for purchasing
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x=sales_gen['Gender'], y=sales_gen['Amount'])
```

From the above graph we can see that most buyers are females as compare to males. And even the puchasing power of female is greater than males.

```python
ax = sns.countplot(data=df, x='Age Group', hue='Gender')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
```

From the above graph we can see that most of the buyers are of age between 26-35 years.

#**State**

```python
df.columns
```

```python
#total number of orders from top 10 states

order_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

plt.figure(figsize=(18,5))
sns.barplot(data=order_state, x='State', y='Orders')
plt.title('Number of orders from 10 states')
```

```python
# total amount/sales from top 10 states

sales_states = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

plt.figure(figsize=(18,5))
sns.barplot(data=sales_states, x='State', y='Amount')
plt.title('Total amount from top 10 states')
```

From the above graph we can see that most of the orders are from uttar pradesh, Maharastra and Karnataka. Respectively but the total sales/amount is from UP, Maharashtra and Karnataka.

## **Marital status**

```python
# checking the maritual status

ax = sns.countplot(x=df['Marital_Status'])

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(8,6)})
sns.barplot(data=sales, x='Marital_Status', y='Amount', hue='Gender')
```

From the above graph we can see that most of the buyers are married females and they have high purchasing power.

##Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data=df, x='Occupation')

for bars in ax.containers:
  ax.bar_label(bars)
```

From the above graph we can see that most buyers are working in IT, Healthcare and Aviation.

##Product Category

```python
plt.figure(figsize=(27,7))
ax = sns.countplot(data=df, x='Product_Category')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
product_category_sales = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

plt.figure(figsize=(25,5))
sns.barplot(data=product_category_sales, x='Product_Category', y='Amount')
```

From the above graph we can see that  most of the selling products are Food, Clothing & Apparel and Electronics & Gadgets.

```python
# checking which to 10 selling product by their product ID

product_id_order = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=product_id_order, x='Product_ID', y='Orders')
```

```python
# Top 10 most sold products

df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

Married women age group 26-35 years from Uttar Pradesh, Maharashtra and Karnataka working in IT, Healthcare and Avition are more likely to buy products from Food, Clothing and Electroniucs categories.
